```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Line Graph

```python
gas = pd.read_csv("gas_prices.csv")
# print(gas)

plt.figure(figsize=(9, 7), dpi=100)

plt.title("Gas prices over Time (in USD)", fontdict={'fontweight': 'bold', 'fontsize': 18, 'fontfamily': 'Ubuntu'})

# plt.plot(gas.Year, gas.USA, 'b.-', label="USA")
# plt.plot(gas["Year"], gas.Canada, 'r.-', label="Canada")
# plt.plot(gas["Year"], gas["Italy"], 'g.-', label="Italy")

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker=".", label=country)
        

plt.xticks(gas.Year[::3].tolist() + [2011])

plt.xlabel('Year')
plt.ylabel('US Dollar')

plt.legend()

plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv("fifa_data.csv")
fifa.head(5)
```

### Histogram

```python
bins = [40, 50, 60, 70, 80, 90, 100]

plt.figure(figsize=(6, 4), dpi=150)

plt.hist(fifa.Overall, bins=bins, color="tomato")

plt.xticks(bins)

plt.ylabel("Number of players")
plt.xlabel("Skill level")
plt.title("FIFA rating 2018")

plt.show()
```

### Pie Charts

```python
left_foot = fifa.loc[fifa['Preferred Foot'] == "Left"].count().iloc[0]
right_foot = fifa.loc[fifa['Preferred Foot'] == "Right"].count().iloc[0]

labels = ["Left", "Right"]
colors = ["olive", "green"]

plt.figure(figsize=(6, 6), dpi=100)

plt.pie([left_foot, right_foot], labels=labels, colors=colors, autopct='%.2f %%')

plt.title("Foot preference FIFA")

plt.show()
```

```python
fifa.Weight = [int(x.strip("lbs")) if type(x) == str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count().iloc[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count().iloc[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count().iloc[0]
heavy = fifa.loc[fifa.Weight >= 200].count().iloc[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ["Under 125", "125-150", "150-175", "175-200", "Over 200"]
explode=[.5, 0.1, 0, 0, .5]

plt.title("Weight distribution of FIFA players (lbs)")

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.7, explode=explode)

plt.show()
```

### Whisker Plot

```python
realmadrid = fifa.loc[fifa.Club == "Real Madrid"]["Overall"]
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]["Overall"]

tick_labels = ["Real Madrid", "Barcelona"]

boxes = plt.boxplot([realmadrid, barcelona], tick_labels=tick_labels, patch_artist=True, meprops={'linewidth': 3, 'color': "blue"})

for box in boxes["boxes"]:
    box.set(color="red", linewidth=2)
    box.set(facecolor="white")

plt.title("Football Team Comparison")
plt.ylabel("FIFA Rating")
plt.xlabel("Team")

plt.show()
```
